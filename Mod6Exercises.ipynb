### Exercise 2:

Using the dataframe from the previous exercise, choose features (columns) to create a linear regression formula to predict sales. Try it with and without the y-intercept. How does it make a difference? Does adding or removing features in your model formula make a difference in the output?

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols('sales ~ location + restaurants + employees', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols('sales ~ restaurants + employees', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols(formula='sales ~ location + restaurants + employees - 1', data=df).fit()
result.summary()
```

```python
result = smf.ols(formula='sales ~ location + restaurants - 1', data=df).fit()
result.summary()
```
